```python
import numpy as np
import math, random
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from mpl_toolkits import mplot3d
```

# Section 1 - Univariate function

```python
# build function data
x=-8.1
xx=[]
yy=[]
for i in range(161):
    x+=0.1
    xx.append(x)
    y=4*x*x
    yy.append(y)
```

```python
plt.plot(xx,yy,'-')
plt.xlabel('x', fontsize=25)
plt.ylabel('y', fontsize=25)
plt.text(-1.0,150, r'$4x^2$', fontsize=35)
```

```python
# build derivative data
x0=8.0
ya=[]
xa=[]
for i in range(60):
    xa.append(x0)
    x=x0-0.05*8*x0
    ya.append(4*x0*x0)
    x0=x
```

```python
plt.plot(xa,ya,'o', color='red')
plt.plot(xx,yy,'-')
plt.xlabel('x', fontsize=25)
plt.ylabel('f(x)', fontsize=25)
plt.text(0,200, r'$4x^2$', fontsize=35)
```
